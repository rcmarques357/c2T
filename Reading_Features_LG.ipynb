```python
X = df_training_candidates_set[feature_cols].copy()
y = df_training_candidates_set[target_col].copy()
groups = df_training_candidates_set["meter_number"].copy()
```

```python
from sklearn.model_selection import GroupShuffleSplit

gss = GroupShuffleSplit(n_splits=1, test_size=0.30, random_state=42)
train_idx, valid_idx = next(gss.split(X, y, groups=groups))

X_train = X.iloc[train_idx].copy()
y_train = y.iloc[train_idx].copy()
X_valid = X.iloc[valid_idx].copy()
y_valid = y.iloc[valid_idx].copy()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_valid_scaled = scaler.transform(X_valid)
X_all_scaled = scaler.transform(X)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(
    penalty="l2",
    C=1.0,
    solver="lbfgs",
    class_weight="balanced",
    max_iter=1000,
    random_state=42,
)

model.fit(X_train_scaled, y_train)
```

```python
valid_pred_proba = model.predict_proba(X_valid_scaled)[:, 1]

threshold = 0.5
valid_pred_label = (valid_pred_proba >= threshold).astype(int)
```

```python
from sklearn.metrics import (
    roc_auc_score,
    average_precision_score,
    classification_report,
)

roc_auc = roc_auc_score(y_valid, valid_pred_proba)
pr_auc = average_precision_score(y_valid, valid_pred_proba)

print(f"\nValidation ROC-AUC: {roc_auc:.4f}")
print(f"Validation PR-AUC : {pr_auc:.4f}")
print("\nClassification report:")
print(classification_report(y_valid, valid_pred_label, digits=4))
```

```python

from sklearn.metrics import (
    confusion_matrix,
    ConfusionMatrixDisplay,
    RocCurveDisplay,
    PrecisionRecallDisplay,
    precision_score,
    recall_score,
    f1_score,
    accuracy_score,
    balanced_accuracy_score,
)
import matplotlib.pyplot as plt

cm = confusion_matrix(y_valid, valid_pred_label)
print("\nConfusion matrix (rows=actual, cols=pred):")
print(cm)

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0, 1])
disp.plot(values_format="d")
plt.title("Confusion Matrix (Current Threshold)")
plt.show()
```

```python
import numpy as np
thresholds = np.linspace(0.05, 0.95, 19)

rows = []
for t in thresholds:
    pred_t = (valid_pred_proba >= t).astype(int)
    rows.append({
        "threshold": t,
        "accuracy": accuracy_score(y_valid, pred_t),
        "balanced_acc": balanced_accuracy_score(y_valid, pred_t),
        "precision_1": precision_score(y_valid, pred_t, pos_label=1, zero_division=0),
        "recall_1": recall_score(y_valid, pred_t, pos_label=1, zero_division=0),
        "f1_1": f1_score(y_valid, pred_t, pos_label=1, zero_division=0),
        "precision_0": precision_score(y_valid, pred_t, pos_label=0, zero_division=0),
        "recall_0": recall_score(y_valid, pred_t, pos_label=0, zero_division=0),
        "f1_0": f1_score(y_valid, pred_t, pos_label=0, zero_division=0),
    })

df_thr = pd.DataFrame(rows).sort_values("f1_1", ascending=False)

print("\nTop thresholds by F1 for class 1 (positive class):")
print(df_thr.head(10).to_string(index=False))

best_t = df_thr.iloc[0]["threshold"]
print(f"\nBest threshold by F1(class=1): {best_t:.2f}")
```

```python

df_eval = pd.DataFrame({
    "y_true": np.asarray(y_valid),
    "proba_1": np.asarray(valid_pred_proba),
})
df_eval["y_pred_0.5"] = (df_eval["proba_1"] >= 0.5).astype(int)
df_eval["error_0.5"] = (df_eval["y_true"] != df_eval["y_pred_0.5"])

print("\nMost confident WRONG predictions (threshold=0.5):")
wrong = df_eval[df_eval["error_0.5"]].copy()
wrong["confidence"] = np.where(wrong["y_pred_0.5"] == 1, wrong["proba_1"], 1 - wrong["proba_1"])
print(wrong.sort_values("confidence", ascending=False).head(15).to_string(index=False))
```

```python
top2_df = (
    df.sort_values(["meter_number", "lr_score_probability"], ascending=[True, False])
      .groupby("meter_number")
      .head(2)
      .copy()
)

top2_df["rank"] = top2_df.groupby("meter_number").cumcount() + 1

top1 = top2_df[top2_df["rank"] == 1].copy()
second = top2_df[top2_df["rank"] == 2][
    ["meter_number", "Candidate_Transformer", "lr_score_probability"]
].copy()

second = second.rename(
    columns={
        "Candidate_Transformer": "second_best_transformer_id",
        "lr_score_probability": "second_best_score",
    }
)

top1 = top1.merge(second, on="meter_number", how="left")
top1["second_best_score"] = top1["second_best_score"].fillna(0.0)
top1["score_gap"] = top1["lr_score_probability"] - top1["second_best_score"]

def classify_assignment(row):
    if row["lr_score_probability"] >= 0.80 and row["score_gap"] >= 0.15:
        return "assigned"
    if row["lr_score_probability"] >= 0.65:
        return "review"
    return "unassigned"

top1["assignment_status"] = top1.apply(classify_assignment, axis=1)

final_assignment = top1[
    [
        "meter_number",
        "Candidate_Transformer",
        "lr_score_probability",
        "second_best_transformer_id",
        "second_best_score",
        "score_gap",
        "assignment_status",
    ]
]

print("\nFinal assignment:")
print(final_assignment)
```
